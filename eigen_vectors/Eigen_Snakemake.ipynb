# Snakemake Example

```python
%%bash
snakemake -s eigen.snakefile
```

In our example Snakemake workflow, we:

1. Create 2D matrix of random numbers and store as 8 distinct NumPy files

2. Find the eigenvalues and eigenvectors of these 8 random matrices

3. Find the sum of the eigenvalues for each file

4. Sort the sums of our 8 tasks and print the output
